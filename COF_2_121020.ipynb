Buat Grid Graph

```
data = pd.DataFrame(data= Data_Kalibrasi_Grade,columns=["SAMPLE_ID,C,10", "CAKA,N,19,4", "FEKA,N,19,4",	"FEKB,N,19,4", 	"NIKA,N,19,4", 	"CUKA,N,19,4", "CUKB,N,19,4",	"PBLA,N,19,4",	"ASKB,N,19,4",	"ULA,N,19,4", "PREATEMP,N,19,4", "Cu_Grade" ])
Grid_Data = sns.PairGrid(data)
Grid_Data.map(plt.scatter)
```

```
data_filter = data[data['PREATEMP,N,19,4'] < 16][data['Cu_Grade'] < 1][data['PREATEMP,N,19,4'] > 1][data["CUKA,N,19,4"] < 110]
Grid_Filter_data = sns.PairGrid(data_filter)
Grid_Filter_data.map(plt.scatter)
```
