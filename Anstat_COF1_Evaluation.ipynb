```
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline 
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats, integrate
sns.set(color_codes=True) 
```

```
Data_Calibration_raw = pd.read_excel('Calibration Data COF 1 Cu.xls')
Data_Calibration_raw.head()
Data_Calibration = Data_Calibration_raw[Data_Calibration_raw['Cu_Grade'] < 1]
Data_Calibration.head()
```

# Grid Data 

```
Grid_Size = sns.PairGrid(Data_Calibration)
Grid_Size.map(plt.scatter);
```

#Connection between FeKa, CuKa, and Cu Grade 

```
Cu_LabGrade= Data_Calibration["Cu_Grade"];CuKa = Data_Calibration['CUKA,N,19,4']; FeKb = Data_Calibration['FEKB,N,19,4']
sns.axes_style('white')
Grade_CuKa = sns.jointplot(data=Data_Calibration, x= Cu_LabGrade, y= CuKa, kind = "reg")
Grade_FeKb = sns.jointplot(data=Data_Calibration, x= Cu_LabGrade, y= FeKb, kind = "reg")
```
