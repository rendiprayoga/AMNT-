```
import matplotlib.pyplot as plt
plt.style.use('classic')
%matplotlib inline 
import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats, integrate
sns.set(color_codes=True) 
```

```
CST_Calibration_Data = pd.read_excel('Calibration Data CST CU Raw.xls')
CST_Calibration_Data.head()
```

```
Grid_Grade = sns.PairGrid(CST_Calibration_Data)
Grid_Grade.map(plt.scatter)
```

```
Cu_LabGrade= CST_Calibration_Data["Cu_Grade "];CuKa = CST_Calibration_Data['CUKA,N,19,4']; CuKb = CST_Calibration_Data['CUKB,N,19,4']; ZnKa = CST_Calibration_Data['ZNKA,N,19,4']
sns.axes_style('white')
Grade_CuKa = sns.jointplot(data=CST_Calibration_Data, x= Cu_LabGrade, y= CuKa, kind = "reg")
Grade_CuKb = sns.jointplot(data=CST_Calibration_Data, x= Cu_LabGrade, y= CuKb, kind = "reg")
Grade_ZnKa = sns.jointplot(data=CST_Calibration_Data, x= Cu_LabGrade, y= ZnKa, kind = "reg")
```
