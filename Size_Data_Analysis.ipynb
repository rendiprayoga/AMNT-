#Import Data 

```
import seaborn as sns, numpy as np
type(data_size_analysis['Feed Anstat1 %Cu'])
COF_BoxCluster1Size = data_size_analysis[102]; COF_BoxCluster2Size = data_size_analysis[103]; COF_BoxCluster3Size = data_size_analysis[203]; COF_BoxCluster4Size = data_size_analysis[204]
COF_Anstat1Size = data_size_analysis["COF 1 Ans Size"]; COF_Anstat2Size = data_size_analysis["COF 2 Ans Size"]
CuGrade101 = data_size_analysis['COF102 %Cu ']; CuGrade102 = data_size_analysis['COF103 %Cu ']; CuGrade201 = data_size_analysis['COF202 %Cu ']; CuGrade202 = data_size_analysis['COF203 %Cu']
CuGradeAnstat1 = data_size_analysis['Feed Anstat1 %Cu']; CuGradeAnstat2 = data_size_analysis['Feed Anstat 2 %Cu']
```

#Plot Sizing 

```
sns.axes_style('white')
Cluster1_Size = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster1Size, y= COF_Anstat1Size, kind = "reg")
Cluster2_Size = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster2Size, y= COF_Anstat1Size, kind = "reg")
Cluster3_Size = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster3Size, y= COF_Anstat2Size, kind = "reg")
Cluster4_Size = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster4Size, y= COF_Anstat2Size, kind = "reg")
Line1_Comparison = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster1Size, y = COF_BoxCluster2Size, kind = "reg")
Line2_Comparison = sns.jointplot(data=data_size_analysis, x= COF_BoxCluster3Size, y = COF_BoxCluster4Size, kind = "reg")
Cluster1_Size.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
Cluster2_Size.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
Cluster3_Size.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
Cluster4_Size.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
Line1_Comparison.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
Line2_Comparison.ax_joint.plot(np.linspace(100, 300),
                         np.linspace(100, 300), ':k')
```
